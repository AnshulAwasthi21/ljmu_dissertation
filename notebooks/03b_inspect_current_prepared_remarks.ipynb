```python
import json
import pandas as pd

# Load the current Phase 2 corpus
df = pd.read_parquet("../data/processed/combined_corpus_n200.parquet")

# Filter to just the surviving prepared_remarks units
prepared = df[df["subtype"] == "prepared_remarks"].reset_index(drop=True)
print(f"Found {len(prepared)} prepared_remarks units")
print()

# Print first 500 characters of each for visual inspection
for i, row in prepared.iterrows():
    # metadata may be JSON-serialized — handle either case
    meta = row["metadata"]
    if isinstance(meta, str):
        meta = json.loads(meta)
    ticker = meta.get("ticker", "?")
    
    print(f"--- Unit {i+1}/{len(prepared)} | doc_id={row['doc_id']} | "
          f"ticker={ticker} | length={row['char_length']} chars ---")
    print(row["text"][:500])
    print()
    print("=" * 80)
    print()
```

After the speakers' presentation, there will be a question and answer session. Please note that today's conference is being recorded. I would now like to hand the call over to..."
